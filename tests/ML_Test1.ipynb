### Exercise 2

Make a set of association rules for support=0.01 and confidence=0.8 add parameter `num_itemsets=5` (if needed).

```python
transactions = []
for i in range(music.shape[0]):
  row=music.iloc[i].dropna().tolist()
  transactions.append(row)

transactions
```

```python
te=TransactionEncoder()
te_ary=te.fit(transactions).transform(transactions)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
frequent_itemsets=apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8, num_itemsets=5)
rules = rules[['antecedents','consequents','support','confidence','lift']]
rules
```

### Exercise 3

Find the set of rules for which consequents is the set `{rammstein}`. Find the rule with the highest support value. Give its interpretation.

```python
rules[rules['consequents']=={'rammstein'}]
```

```python
rammstein_consq=rules[rules['consequents']=={'rammstein'}]
rammstein_consq.loc[rammstein_consq['support'].idxmax()]
```

```python
print(f"We can see that the highest support value (in this sub-set) is {rammstein_consq['support'].max()}")
```

### Exercise 4

Which performer/band is the most frequent in this dataset

```python
frequent_itemsets=apriori(df, min_support=0.01, use_colnames=True)
performer=frequent_itemsets.sort_values(by='support', ascending=False).head(1)
performer
```

```python
print(f"The most frequent performer/band is {performer['itemsets'].iloc[0]}")
```
