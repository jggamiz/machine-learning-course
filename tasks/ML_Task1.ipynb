```python
import pandas as pd
```

```python
df = pd.read_csv("./pets.csv")
```

```python
#print("Number of colums: ", df.columns.size) #this also works
print("Number of colums: ", df.shape[1])
print("Number of rows: ", df.shape[0])
```

```python
print("Numerical values:")
print(df.describe().head(0))
print("\n")
print("Qualitative variables:")
print(df.describe(include='object').head(0))
```

```python
cats_df = df[df['pet'] == 'cat']
young_cats_df = cats_df[cats_df['age'] < 5]
mean_rating = df['score'].mean()
high_rated_young_cats_df = young_cats_df[young_cats_df['score'] > mean_rating]
#print(high_rated_young_cats_df)
pd.DataFrame(high_rated_young_cats_df, columns=["id", "pet", "country", "score", "age", "weight"])
```

```python
cross_tab = pd.crosstab(df['pet'], df['country'])
#print(cross_tab)
pd.DataFrame(cross_tab)
```
