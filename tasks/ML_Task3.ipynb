```python
movies = pd.read_csv("/content/movie_dataset.csv", header=None)
movies
```

```python
movies.shape
```

```python
transactions = []
for i in range(movies.shape[0]):
  row=movies.iloc[i].dropna().tolist()
  transactions.append(row)

transactions
```

```python
len(transactions)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
```

Let's see some rules:

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.25, num_itemsets=len(frequent_itemsets))
rules = rules[['antecedents','consequents','support','confidence','lift']]
rules
```

Now we find the highest value of the support, and identify the rule associated with that value

```python
rules.loc[rules['support'].idxmax()]
```

```python
rules['support'].max()
```
