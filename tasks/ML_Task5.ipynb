```python
df  = pd.read_csv('/content/basic1.csv')
df
```

```python
df.color.value_counts()
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='color')
```

```python
from sklearn.preprocessing import StandardScaler

data_scaled = StandardScaler().fit_transform(df[['x','y']])
data_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
num_clusters = 4

km = KMeans(n_clusters=num_clusters, random_state=1)
km.fit(data_scaled)

# Predict the target variable
pred = km.predict(data_scaled)
pred
```

```python
sns.scatterplot(data=df, x='x', y='y', hue=pred, palette='bright')
```
