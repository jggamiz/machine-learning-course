```python
X = voice.drop('label', axis=1)
y = voice['label']
```

## kNN

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train.value_counts(normalize=True) # to check there's almost the same proportion
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

First, we have to find the best value for $k$

```python
error1 = []
error2 = []

for k in range(1,21,2):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  y_pred1 = knn.predict(X_train)
  error1.append(np.mean(y_pred1!=y_train))
  y_pred2 = knn.predict(X_test)
  error2.append(np.mean(y_pred2!=y_test))
```

```python
if 'min' in globals():
    del min

min_score = min(error2)
min_index = error2.index(min_score)
min_index+1
```

```python
plt.plot(range(1,21,2), error1, label='train')
plt.plot(range(1,21,2), error2, label='test')
plt.legend()
plt.xlabel('k')
plt.ylabel('Error')
```

```python
accuracy_rate = []

for k in range(1,11,2):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  score = accuracy_score(y_test, y_pred)
  accuracy_rate.append(score)
```

```python
plt.plot(range(1,11,2), accuracy_rate)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.legend()
```

We conclude that the best option is $k=3$

```python
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
