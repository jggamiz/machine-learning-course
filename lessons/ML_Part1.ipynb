### Handling outliers

Outliers are those data points that are distant from most of the similar points – in other words, we can say the outliers are entities that are different from the crowd.

Outliers cause problems when it comes to building predictive models, such as long model training times, poor accuracy, an increase in error variance, a decrease in normality, and a reduction in the power of statistical tests.



```python
q1 = np.quantile(house.Price, 0.25)
q1
```

```python
q3 = np.quantile(house.Price, 0.75)
q3
```

```python
iqr = q3-q1
iqr
```

```python
upper_lim = q3+(1.5*iqr)
upper_lim
```

```python
lower_lim = q1-(1.5*iqr)
lower_lim
```

```python
outliers = house[(house['Price'] > upper_lim) | (house['Price'] < lower_lim)]
outliers
```

#### Frequent itemsets

The apriori function expects data in a one-hot encoded pandas dataFrame.

```python
'''
TransactionEncoder converts a list of transactions into a one-hot encoded DataFrame or array, where:

 - Rows represent individual transactions.
 - Columns represent items across all transactions.
 - Cells contain binary values (1 or 0), indicating the presence or absence of an item in a particular transaction.
'''

from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

#### Association rules from Frequent Itemsets

The generate_rules takes dataframes of frequent itemsets as produced by the `apriori`, `fpgrowth`, or `fpmax` functions in mlxtend.association.

The `generate_rules()` function allows you to

1.   specify your metric of interest
2.   the according threshold


```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4, num_itemsets=len(frequent_itemsets))
```

For the rule $antecedents \Rightarrow consequents$
$$
\left(A\Rightarrow C\right)
$$
we have



*  the **support** is used to measure the abundance or frequency (often interpreted as significance or importance) of an itemset in a database.
$$
range:[0,1], \\ support(A⇒C)=\dfrac{σ(A\cup B)}{N}
$$
*  the **confidence** of a rule $A\Rightarrow C$ is the probability of seeing the consequent in a transaction given that it also contains the antecedent.
$$
range:[0,1], \\ confidence(A⇒C) = \dfrac{σ(A\cup C)}{σ(A)}
$$
*  the **lift** metric is commonly used to measure how much more often the antecedent and consequent of a rule $A\Rightarrow C$ occur together than we would expect if they were statistically independent. If $A$ and $C$ are independent, the Lift score will be exactly $1$.
$$
range:[0,\infty], \\ confidence(A⇒C) = \dfrac{σ(A\cup C)\cdot N}{σ(A)σ(C)}
$$
* the leverage computes the difference between the observed frequency of $A$ and $C$ appearing together and the frequency that would be expected if $A$ and $C$ were independent. A leverage value of $0$ indicates independence.
$$
range:[-1,1]
$$
*  A high conviction value means that the consequent is highly depending on the antecedent. If items are independent, the conviction is $1$.
$$
range:[0,\infty]
$$
*  the zhangs_metric measures both association and dissociation. Value ranges between $-1$ and $1$. A positive value ($>0$) indicates Association and negative value indicated dissociation.
$$
range:[-1,1]
$$

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.14, num_itemsets=len(frequent_itemsets))
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules
```

```python
rules[ (rules['confidence']>0.35) & (rules['lift']>1.2) ]
```

```python
rules[rules['antecedents'] == {'MILK'}]
```

```python
rules[rules['antecedents'] == {'BREAD'}]
```
